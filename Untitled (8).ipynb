```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
diabetes_data = load_diabetes()

df = pd.DataFrame(data=diabetes_data.data, columns=diabetes_data.feature_names)
df['target'] = diabetes_data.target

df.head()
```

```python
df.isnull().sum()
```

```python
scaler = StandardScaler()
X = df.drop('target', axis=1)
y = df['target']

X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(y, kde=True)
plt.title('Distribution of Diabetes Progression')
plt.xlabel('Diabetes Progression (target)')
plt.ylabel('Frequency')
plt.show()

df.drop(columns='target').hist(figsize=(12, 10), bins=15)
plt.suptitle('Feature Distributions')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

sns.scatterplot(x=df['bmi'], y=df['target'])
plt.title('BMI vs Diabetes Progression')
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(64, input_dim=X_scaled.shape[1], activation='relu'))

model.add(Dense(32, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

model.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test), verbose=1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
loss, mae = model.evaluate(X_test, y_test)

print(f'Mean Squared Error (MSE): {loss:.2f}')
print(f'Mean Absolute Error (MAE): {mae:.2f}')
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f'R² Score: {r2:.2f}')
```

```python
improved_model = Sequential()
improved_model.add(Dense(128, input_dim=X_scaled.shape[1], activation='relu'))
improved_model.add(Dense(64, activation='relu'))
improved_model.add(Dense(32, activation='relu'))
improved_model.add(Dense(1))
improved_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

improved_history = improved_model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test), verbose=1)

plt.plot(improved_history.history['loss'], label='Training Loss')
plt.plot(improved_history.history['val_loss'], label='Validation Loss')
plt.title('Improved Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
improved_loss, improved_mae = improved_model.evaluate(X_test, y_test)
print(f'Improved Model MSE: {improved_loss:.2f}')
print(f'Improved Model MAE: {improved_mae:.2f}')
y_pred_improved = improved_model.predict(X_test)
improved_r2 = r2_score(y_test, y_pred_improved)
print(f'Improved R² Score: {improved_r2:.2f}')
```
